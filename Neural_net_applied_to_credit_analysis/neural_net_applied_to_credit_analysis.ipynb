```python
import pickle
with open('credit.pkl','rb') as f:
    X_train, y_train, X_test, y_test = pickle.load(f)
```

```python
X_train.shape, y_train.shape
```

```python
X_test.shape, y_test.shape
```

```python
from sklearn.neural_network import MLPClassifier
net_credit = MLPClassifier(max_iter=1800, verbose=True, tol=0.00001, solver = 'adam',
                           activation= 'relu', hidden_layer_sizes = (2,2))
net_credit.fit(X_train, y_train)
```

```python
predictions = net_credit.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report
print('The accuracy of this neuralnet predictions is of:',accuracy_score(y_test, predictions)*100,'%')
```

```python
print(classification_report(y_test, predictions))
```

```python
from yellowbrick.classifier import ConfusionMatrix
cm = ConfusionMatrix(net_credit)
cm.fit(X_train,y_train)
cm.score(X_test,y_test)
```
